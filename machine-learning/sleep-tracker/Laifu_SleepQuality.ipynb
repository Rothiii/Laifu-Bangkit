# Data Preprocessing

```python
#Drop Person ID Column and Empty Column
df.drop(columns=['Person ID'], inplace=True)
df.dropna(inplace=True)
```

```python
#Replace Gender into Female and Male
df['Gender'].replace({'Female':0,'Male':1},inplace=True)

#Replace BMI Category into List of Number
df['BMI Category'].replace({'Normal':0,'Normal Weight':0,'Overweight':1,'Obese':2},inplace=True)

#Replace Occupation into List of Number
unique_occupation = df['Occupation'].unique()
df['Occupation'].replace(to_replace=unique_occupation, value=list(range(len(unique_occupation))), inplace=True)

#Split Blood Pressure Column into Sistole and Diastole
df[['Sistole', 'Diastole']] = df['Blood Pressure'].str.split('/', expand=True)
df[['Sistole', 'Diastole', 'Daily Steps', 'Physical Activity Level']] = df[['Sistole', 'Diastole', 'Daily Steps', 'Physical Activity Level']].astype(np.int64)
df.drop(columns=['Blood Pressure'], inplace=True)
```

```python
x = df.drop(columns=['Sleep Quality'])
y = df['Sleep Quality']
```

```python
train_size = int(0.8 * x.shape[0])

x_train = x[:train_size]
y_train = y[:train_size]

x_valid = x[train_size:]
y_valid = y[train_size:]
```

```python
x_train.dtypes
```

```python
x_train.head(10)
```

```python
tf_dataset = tf.data.Dataset.from_tensor_slices((x_train.values, y_train.values)).batch(32)
tf_dataset_val = tf.data.Dataset.from_tensor_slices((x_valid.values, y_valid.values)).batch(32)
```

```python
for features, targets in tf_dataset.take(1):
  print ('Features: {}, Target: {}'.format(features, targets))
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(32, activation = 'relu', input_shape = (11, )),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(1)
])

model.summary()

model.compile(optimizer = tf.keras.optimizers.Adam(), loss = 'mae', metrics = ['mae'])

model.fit(tf_dataset, epochs = 10, verbose = 1, validation_data = tf_dataset_val)
```
